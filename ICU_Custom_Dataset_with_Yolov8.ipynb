```python
import yaml

data = {
    "train" : "/content/ICU_dataset-10/train/images/",
    "val" : "/content/ICU_dataset-10/valid/images",
    "test" : "/content/ICU_dataset-10/test/images",
    "names":["fist","hammer","knife"],
    "nc":3
}
# detection 하고 싶은 클래스 개수와 클래스에 대응되는 클래스 이름

with open("/content/ICU_dataset-10/data.yaml", "w") as f:
  yaml.dump(data, f)

with open("/content/ICU_dataset-10/data.yaml", "r") as f:
  icu_yaml = yaml.safe_load(f)
  display(icu_yaml)
```

```python
from ultralytics import YOLO
# pre-trained model
model = YOLO("yolov8n.pt") #작은 걸로 학습, ms coco dataset으로 사전학습된거.
```

```python
print(type(model.names), len(model.names))
print(model.names)
```

```python
# train the model (custom data.yaml 참조하여 학습 : fine-tuning)
model.train(data="/content/ICU_dataset-10/data.yaml", epochs=50, patience=30, batch=32, imgsz=640)
```

```python
results = model.val()  # evaluate model performance on the validation set
```

```python
results = model.predict(source="/content/ICU_dataset-10/test/images", save=True,conf = 0.621 )
# ms coco data로 사전학습되어 있기에 ms coco dataset에 정의되어 있는 클래스 개수와 종류
# 학습된 결과의 클래스 이름, 개수는 다름
```
